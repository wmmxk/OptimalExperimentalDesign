```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
import numpy as np
import random

```

```python
def gen_sur(X,Y,random_seed=2017,sigma=0.5):
    random.seed(random_seed)
    for i in range(300):
        mu1,mu2 = np.array([random.uniform(-10,10) for i in range(2)])
        sigma1 = random.uniform(sigma,3)
        sigma2 = random.uniform(sigma*0.8,4)
        sigma12 = random.uniform(min(sigma1,sigma2)/20,min(sigma1,sigma2)/2)
        if i ==0:
            Z = mlab.bivariate_normal(X,Y,sigma1, sigma2,mu1,mu2,sigma12)
        else:
            Z = Z + mlab.bivariate_normal(X,Y,sigma1, sigma2,mu1,mu2,sigma12)
    return Z
```

```python
#from pylab import *
fig = plt.figure()
ax = fig.gca(projection = '3d')
rainbow = plt.get_cmap('rainbow')

x = np.linspace(-10, 10, 120)
y = np.linspace(-10, 10, 120)
X, Y = np.meshgrid(x, y)

Z = gen_sur(X,Y)

surf = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = rainbow, linewidth = 0)
ax.set_zlim3d(0, Z.max())

plt.show()

fig.savefig('./out_fig/test.pdf',bbox_inches = 'tight')
```
