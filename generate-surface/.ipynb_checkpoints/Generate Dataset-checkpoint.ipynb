```python
import numpy as np

def f1(f, phi,d=0.1,k=5):
    speed_of_sound = 330
    summed = 1
    for n in np.arange(1, (k-1)/2 + 1, 1):
        summed += 2 * np.cos(2 * np.pi * f / speed_of_sound * n * d * np.cos(phi))
    out = np.abs(summed / np.max(summed))
    return out
```

```python
import numpy as np
class GenerateData(object):
    def __init__(self,f,boundaries):
        self.f = f
        self.xbs = boundaries[0]
        self.ybs = boundaries[1]
    def create(self):
        xs = np.linspace(self.xbs[0], self.xbs[1],70)
        ys = np.linspace(self.ybs[0], self.ybs[1],70)
        x, y = np.meshgrid(xs,ys)
        z = self.f(x,y)
        data = np.stack((x,y,z),axis=0)
        d3 = data.reshape((3,-1))
        d3 = d3.transpose((1,0))
        return d3
```

```python
generateData = GenerateData(f1,[(0,5000),(0,3.142)])
d3 = generateData.create()
```

```python

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline



fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(d3[:,0], d3[:,1], d3[:,2], linewidth=0.2, antialiased=True)
#ax.scatter3D(x, y, z,c='r')
plt.show()
```
