```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D     # Import 3D plotting tools.
from scipy.special import jn                # Import Bessel function.
from matplotlib.colors import LightSource

# Get colormaps to use with lighting object.
from matplotlib import cm


# Define grid of points.
points = np.linspace(-10, 10, 51)
X, Y = np.meshgrid(points, points)
R = np.sqrt(X**2 + Y**2)
Z = jn(0,R)

x,y = 6,0
p = np.sqrt(x**2 + y**2)
z = jn(0,p)


light = LightSource(90, 45)
illuminated_surface = light.shade(Z, cmap=cm.coolwarm)

rgb = np.ones((Z.shape[0], Z.shape[1], 3))
green = np.array([247,236,156])  # This color is borrowed from dribbble.com; the hex code is converted to rgb
green_surface = light.shade_rgb(rgb * green, Z)

fig = plt.figure(figsize=(8,5))
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False,
                facecolors=green_surface)

ax.view_init(elev=50,azim=45) # azim is the Azimuth angle in x-y plane
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

xy_ticks = np.linspace(-10, 10, 5)
z_ticks = np.linspace(-0.4, 1, 5)
ax.set_xticks(xy_ticks, minor = False)
ax.set_yticks(xy_ticks, minor = False)
ax.set_zticks(z_ticks, minor = False)

ax.plot([x],[y],[z],marker="*")

fig.savefig('out_fig/test.pdf',bbox_inches = 'tight')
```
