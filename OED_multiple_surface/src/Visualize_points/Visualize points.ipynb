```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pylab as pl

fig, ax = plt.subplots()
fig.set_tight_layout(True)

# Query the figure's on-screen size and DPI. Note that when saving the figure to
# a file, we need to provide a DPI for that separately.
print('fig size: {0} DPI, size in inches {1}'.format(
    fig.get_dpi(), fig.get_size_inches()))

# Plot a scatter that persists (isn't redrawn) and the initial line.

ax.scatter(points.iloc[:50,1],points.iloc[:50,2],s=3)

def update(i):
    ax.scatter(points.iloc[50+i-3:50+i,1],points.iloc[50+i-3:50+i,2],marker= "*")
    for i in range(50+i-3,50+i):
        ax.annotate(i,(points.iloc[i,1], points.iloc[i,2]),fontsize=6)
    return ax


anim = FuncAnimation(fig, update, frames=np.arange(3, 123,3), interval=5000)
anim.save('../../out_fig/line.gif', dpi=250, writer='imagemagick')
```
