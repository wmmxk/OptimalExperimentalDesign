```python

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline
```

```python

from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import pandas as pd
df = pd.read_csv('../../data/surface_real.csv').values
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(df[:,0], df[:,1], df[:,2], linewidth=0.2, antialiased=True,cmap='coolwarm')

plt.show()
```

```python

 
# Make the plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(df[:,0], df[:,1], df[:,2], cmap=plt.cm.viridis, linewidth=0.2)

 
# to Add a color bar which maps values to colors.
surf=ax.plot_trisurf(df[:,0], df[:,1], df[:,2], cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)

 
# Rotate it
ax.view_init(30, 45)

 
# Other palette
ax.plot_trisurf(df[:,0], df[:,1], df[:,2], cmap=plt.cm.jet, linewidth=0.01)
plt.savefig("test.pdf")
```

```python

 
# We are going to do 20 plots, for 20 different angles
for angle in range(70,210,2):
 
    # Make the plot
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot_trisurf(df[:,0], df[:,1], df[:,2],  cmap=plt.cm.jet, linewidth=0.01)

    # Set the angle of the camera
    ax.view_init(30,angle)

    # Save it
    filename='./Volcano_step'+str(angle)+'.png'
    plt.savefig(filename, dpi=296)
    plt.gca()

```
